```python
from kubernetes import client, config
from kubernetes.client import configuration
```

```python
config.load_kube_config(context='default')
```

```python
contexts, active_context = config.list_kube_config_contexts()
print(type(contexts[0]))
contexts = [context['name'] for context in contexts] # context 이름만 갖고 오기 위함 ex) contexts = ['default', 'docker-desktop']

active_context = active_context['name'] 
```

```python
client1 = client.CoreV1Api(api_client=config.new_client_from_config(context=contexts[0]))


print("\nList of pods on %s:" % contexts[0])
for i in client1.list_pod_for_all_namespaces().items:
    print(f"{i.status.pod_ip}\t{i.metadata.namespace}\t{i.metadata.name}")


client2 = client.CoreV1Api(api_client=config.new_client_from_config(context=contexts[1]))

print("\nList of pods on %s:" % contexts[1])
for i in client1.list_pod_for_all_namespaces().items:
    print(f"{i.status.pod_ip}\t{i.metadata.namespace}\t{i.metadata.name}")
```
